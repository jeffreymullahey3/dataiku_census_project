```python
# go ahead and use whole learn data to build a decision tree
```

```python
input_variables = ['has_capital_gains',
'has_capital_losses',
'high_paying_vet_benefits',
'high_paying_own_business_self_employed',
'high_paying_citizenship',
'high_paying_birth_self_country',
'high_paying_birth_mother_country',
'high_paying_birth_father_country',
'high_paying_hhld_summary',
'high_paying_household_and_family_stat',
'high_paying_tax_filer_stat',
'high_paying_employment_stat',
'high_paying_sex',
'high_paying_occupation_code',
'high_paying_industry_code',
'high_paying_marital_stat',
'high_paying_education',
'high_paying_occupation_recode',
'high_paying_industry_recode',
'high_paying_workclass',
'has_dividends_from_stocks',
'weeks_worked_in_year',
'num_persons_worked_for_employer',
'wage_per_hour',
'age']
```

```python
X = df[input_variables]
```

```python
X.head()
```

```python
X.dtypes
```

```python
y = df['income_flag']
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    test_size=0.3,
                                                    stratify=y,
                                                    random_state=42)
```

```python
clf_dt = DecisionTreeClassifier(random_state=42)
clf_dt = clf_dt.fit(X_train, y_train)
```

```python
plot_confusion_matrix(clf_dt, X_test, y_test, display_labels=["<50k", "50k+"])
```

```python
path = clf_dt.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas = path.ccp_alphas
ccp_alphas = ccp_alphas[:-1]

clf_dts = []

for ccp_alpha in ccp_alphas:
    clf_dt = DecisionTreeClassifier(random_state=0, ccp_alpha=ccp_alpha)
    clf_dt.fit(X_train, y_train)
    clf_dts.append(clf_dt)
```
